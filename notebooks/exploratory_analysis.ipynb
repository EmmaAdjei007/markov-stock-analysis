```python


# Add the project root directory to the Python path
sys.path.append("C:/Users/Emmanuel/Documents/Personal/Markov Chain/stock-markov-analysis")

# Import custom modules
from src.data_fetcher import get_stock_data
from src.data_preprocessor import assign_states

# Configure logging
logging.basicConfig(level=logging.INFO, format="%(asctime)s - %(levelname)s - %(message)s")
```

```python

# Constants
TICKERS = ["AAPL", "MSFT", "GOOGL", "AMZN"]
START_DATE = "2021-01-01"
END_DATE = "2024-01-01"
N_STATES = 5
PLOT_COLORS = ['blue', 'green', 'red', 'purple']

# Fetch and process data
logging.info("Fetching stock data...")
raw_data = get_stock_data(TICKERS, START_DATE, END_DATE)

```

```python
if not raw_data:
    logging.error("No data fetched. Check your internet connection or ticker symbols.")
else:
    # Create summary tables
    for ticker, data in raw_data.items():
        if data.empty:
            logging.warning(f"No data found for {ticker}.")
        else:
            print(f"\n📊 Summary for {ticker}:\n")
            print(tabulate(data.describe(), headers="keys", tablefmt="double_grid"))

    # Process data
    logging.info("Processing data and assigning states...")
    processed_data = assign_states(raw_data, n_states=N_STATES)

    # Create combined closing prices plot
    plt.figure(figsize=(14, 8))
    for idx, (ticker, data) in enumerate(processed_data.items()):
        plt.plot(data.index, data['Close'], 
                label=f'{ticker} Close Price', 
                color=PLOT_COLORS[idx % len(PLOT_COLORS)],
                alpha=0.7)
    plt.title(f'Closing Prices Comparison ({START_DATE} to {END_DATE})')
    plt.xlabel('Date')
    plt.ylabel('Price ($)')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Create subplots for individual ticker analysis
    for ticker, data in processed_data.items():
        print(f"\n🔍 Detailed Analysis for {ticker}:\n")
        
        # Create subplot grid
        fig, axs = plt.subplots(3, 1, figsize=(14, 18))
        
        # Closing Price and Rolling Statistics
        data['30D_MA'] = data['Close'].rolling(window=30).mean()
        data['30D_STD'] = data['Close'].rolling(window=30).std()
        
        axs[0].plot(data.index, data['Close'], label='Close Price', color='blue')
        axs[0].plot(data.index, data['30D_MA'], label='30D Moving Average', color='orange')
        axs[0].fill_between(data.index, 
                           data['30D_MA'] - data['30D_STD'], 
                           data['30D_MA'] + data['30D_STD'],
                           color='gray', alpha=0.2)
        axs[0].set_title(f'{ticker} Price Analysis')
        axs[0].set_ylabel('Price ($)')
        axs[0].legend()
        axs[0].grid(True)

        # Daily Returns
        axs[1].plot(data.index, data['Daily_Return'], 
                   color='green', alpha=0.7)
        axs[1].set_title('Daily Returns')
        axs[1].set_ylabel('Return (%)')
        axs[1].grid(True)

        # State Distribution
        state_order = ['Very Bearish', 'Bearish', 'Neutral', 'Bullish', 'Very Bullish']
        sns.countplot(x='State_Description', data=data, 
                     order=state_order, ax=axs[2],
                     palette='viridis')
        axs[2].set_title('State Distribution')
        axs[2].set_xlabel('Market State')
        axs[2].set_ylabel('Count')
        axs[2].tick_params(axis='x', rotation=45)

        plt.tight_layout()
        plt.savefig(f'reports/figures/{ticker}_analysis.png', bbox_inches='tight')
        plt.show()

    # Combined State Distribution Plot
    plt.figure(figsize=(14, 8))
    for idx, (ticker, data) in enumerate(processed_data.items()):
        plt.subplot(2, 2, idx+1)
        sns.countplot(x='State_Description', data=data, 
                     order=state_order, 
                     palette='viridis')
        plt.title(f'{ticker} State Distribution')
        plt.xlabel('')
        plt.ylabel('Count')
        plt.xticks(rotation=45)
    plt.tight_layout()
    plt.savefig('reports/figures/combined_state_distributions.png', bbox_inches='tight')
    plt.show()

    # Combined Returns Distribution
    plt.figure(figsize=(12, 6))
    for idx, (ticker, data) in enumerate(processed_data.items()):
        sns.kdeplot(data['Daily_Return'], 
                   label=ticker, 
                   color=PLOT_COLORS[idx % len(PLOT_COLORS)],
                   fill=True)
    plt.title('Daily Returns Distribution Comparison')
    plt.xlabel('Daily Return (%)')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig('reports/figures/returns_distribution_comparison.png', bbox_inches='tight')
    plt.show()
```
